````python
import pandas as pd
import glob
files=glob.glob("/content/drive/MyDrive/premier_project/csvs_per_year/csvs_per_year**/*.csv")
````

````python
df=pd.DataFrame()
for i in files:
  csv=pd.read_csv(i, parse_dates=["date"],index_col="date")
  df=df.append(csv)
````

````python
df.info()
````

````python
df["CO"]=df["CO"]/1000
df["TCH"]=df["TCH"]/1000
df["CH4"]=df["CH4"]/1000
````

# Cleaning data from "Barrio del pillar","Vallecas",and "Mendez Alvaro"

````python
#group df by station
df_stations= df.groupby(["station"])

#selecting "Barrio del pillar" station from the grouped data
df_barrio = df_stations.get_group(28079039)
df_barrio
````

````python
#checking for missing values for each column in our station 
df_barrio.isnull().sum()/148824*100
````


We decided on filling in misssing values with the median for columns with greater than 1% and less than 70% of missing values.For columns with less than 1% we will in with 0.We will drop columns with more than 70% of misiing values.
```
# This is formatted as code
```



````python
#filling the missing values.
df_barrio["CO"].fillna(0, inplace= True)
df_barrio["NO_2"].fillna(0, inplace=True)
df_barrio["NOx"].fillna(df_barrio["NOx"].median(), inplace=True)
df_barrio["O_3"].fillna(df_barrio["O_3"].median(),inplace=True)
df_barrio["PM10"].fillna(df_barrio["PM10"].median(), inplace=True)
df_barrio["SO_2"].fillna(df_barrio["SO_2"].median(), inplace=True)
df_barrio["NO"].fillna(df_barrio["NO"].median(), inplace=True)
````

````python
df_barrio.drop(["BEN","EBE","MXY","NMHC","OXY","PXY","TCH","TOL","PM25","CH4"],axis=1)
````

````python
#checking for missing values
df_barrio.isnull().sum()
````

````python
#drops columns with more than 70% missing values
 df_barrio=df_barrio.drop(["BEN","EBE","MXY","NMHC","OXY","PXY","TCH","TOL","PM25","CH4"],axis=1)
````

````python
df_barrio.isnull().sum()
````

````python
#Checking for duplicate data in our data frame
sum(df_barrio.duplicated())
````

````python
#dropping all duplicates
df_barrio.drop_duplicates(inplace=True)
````

````python
#checking if duplicate problem solved 
sum(df_barrio.duplicated())
````

````python
df_barrio.to_csv("/content/drive/MyDrive/premier_project/cleaned_data/barrio_del_pillar.csv",index=False)
````

````python
#station Vallecas
df_stations= df.groupby(["station"])
df_vallecas= df_stations.get_group(28079040)
df_vallecas
````

````python
#% of missing values in each column in vallecas data frame
df_vallecas.isnull().sum()/151608*100
````

````python
#solving for missing values by filling with the median for below 70% and above 1%,filling with 0 for missing values below 1%
df_vallecas["CO"].fillna(df_vallecas["CO"].median(), inplace= True)
df_vallecas["NO_2"].fillna(0, inplace=True)
df_vallecas["NOx"].fillna(df_vallecas["NOx"].median(), inplace=True)
df_vallecas["O_3"].fillna(df_vallecas["O_3"].median(),inplace=True)
df_vallecas["PM10"].fillna(0, inplace=True)
df_vallecas["SO_2"].fillna(0, inplace=True)
df_vallecas["NO"].fillna(df_vallecas["NO"].median(), inplace=True)
````

````python
#dropping all columns with more than 70% missing values
df_vallecas=df_vallecas.drop(["BEN","EBE","MXY","NMHC","OXY","PXY","TCH","TOL","PM25","CH4"],axis=1)
````

````python
#sum of missing values in each column
df_vallecas.isnull().sum()
````

````python
#checking if the dataframe has any duplicates
sum(df_vallecas.duplicated())
````

````python
 #dropping all duplicates in the dataframe
df_vallecas.drop_duplicates(inplace=True)
````

````python
#finding out if any duplicates are left
sum(df_vallecas.duplicated())
````

````python
df_vallecas.to_csv("/content/drive/MyDrive/premier_project/cleaned_data/vallecas.csv",index=False)
````
