```python
import pandas as pd 
import numpy as np
```

```python
madrid_1 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2001.csv")
madrid_2 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2002.csv")
madrid_3 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2003.csv")
madrid_4 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2004.csv")
madrid_5 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2005.csv")
madrid_6 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2006.csv")
madrid_7 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2007.csv")
madrid_8 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2008.csv")
madrid_9 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2009.csv")
madrid_10 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2010.csv")
madrid_11 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2011.csv")
madrid_12 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2012.csv")
madrid_13 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2013.csv")
madrid_14 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2014.csv")
madrid_15 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2015.csv")
madrid_16 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2016.csv")
madrid_17 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2017.csv")
madrid_18 = pd.read_csv("./csvs_per_year/csvs_per_year/madrid_2018.csv")
```

```python
total_df = pd.concat([madrid_1, madrid_2, madrid_3, madrid_4, madrid_5, madrid_6,
          madrid_7, madrid_8, madrid_9, madrid_10, madrid_11, madrid_11, madrid_12,
          madrid_13, madrid_14, madrid_15, madrid_16, madrid_17, madrid_18], axis=0)
```

```python
total_df["CO"]=total_df["CO"]/1000
total_df["TCH"]=total_df["TCH"]/1000
total_df["CH4"]=total_df["CH4"]/1000
```

```python
arturo_soria = total_df.loc[total_df["station"] == 28079016].reset_index(drop=True)
villaverde = total_df.loc[total_df["station"] == 28079017].reset_index(drop=True)
farolillo = total_df.loc[total_df["station"] == 28079018].reset_index(drop=True)
```

```python
## Conversion of date to pandas timestamp

arturo_soria["date"] = pd.to_datetime(arturo_soria["date"])
villaverde["date"] = pd.to_datetime(villaverde["date"])
farolillo["date"] = pd.to_datetime(farolillo["date"])
```

```python
pd.DataFrame(arturo_soria.isna().sum()/arturo_soria.shape[0]*100)
```

Dropping the columns with more than 70% missing values

```python
arturo_soria.drop(["BEN", "EBE", "MXY", "NMHC", "OXY", "PXY", "TCH", "TOL", "PM25", "CH4"], axis=1, inplace=True)
```

```python
## Missing values less than 1%
arturo_soria["CO"].fillna(0, inplace=True)
arturo_soria["NO_2"].fillna(0, inplace=True)
arturo_soria["O_3"].fillna(0, inplace=True)

## Missing values are more than 1% and less than 70%
arturo_soria["NOx"].fillna(arturo_soria["NOx"].median(), inplace=True)
arturo_soria["PM10"].fillna(arturo_soria["PM10"].median(), inplace=True)
arturo_soria["SO_2"].fillna(arturo_soria["SO_2"].median(), inplace=True)
arturo_soria["NO"].fillna(arturo_soria["NO"].median(), inplace=True)
```

```python
## Check for duplicates
arturo_soria.duplicated().sum()
```

```python
arturo_soria.drop_duplicates(inplace=True)
```

```python
arturo_soria.to_csv("arturo_soria.csv", index=False)
```

### farolillo data cleaning

```python
pd.DataFrame(farolillo.isna().sum()/farolillo.shape[0]*100)
```

```python
## Drop columns with missing values more than 70%
farolillo.drop(["CH4", "PM25", "TCH", "PXY", "OXY", "NMHC", "MXY"], axis=1, inplace=True)

## Filling columns with missing values less than 1% using zero(0)
farolillo["SO_2"].fillna(0, inplace=True)
farolillo["O_3"].fillna(0, inplace=True)
farolillo["NO_2"].fillna(0, inplace=True)
farolillo["CO"].fillna(0, inplace=True)

## Filling columns with missing values greater than 1% and less than 70% using median values 
farolillo["NO"].fillna(farolillo["NO"].median(), inplace=True)
farolillo["TOL"].fillna(farolillo["TOL"].median(), inplace=True)
farolillo["PM10"].fillna(farolillo["PM10"].median(), inplace=True)
farolillo["NOx"].fillna(farolillo["NOx"].median(), inplace=True)
farolillo["EBE"].fillna(farolillo["EBE"].median(), inplace=True)
farolillo["BEN"].fillna(farolillo["BEN"].median(), inplace=True)
```

```python
## Drop duplicates 
print(farolillo.duplicated().sum())
farolillo.drop_duplicates(inplace=True)
```

```python
farolillo.to_csv("farolillo.csv", index=False)
```
