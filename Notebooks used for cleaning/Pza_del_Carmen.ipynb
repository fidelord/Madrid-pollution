```python
import pandas as pd
```

```python
madrid =['madrid_2001.csv','madrid_2002.csv',
         'madrid_2003.csv','madrid_2004.csv','madrid_2005.csv','madrid_2006.csv',
        'madrid_2007.csv','madrid_2008.csv','madrid_2009.csv','madrid_2010.csv',
        'madrid_2011.csv','madrid_2012.csv','madrid_2013.csv','madrid_2014.csv',
        'madrid_2015.csv','madrid_2016.csv','madrid_2017.csv','madrid_2018.csv']
```

```python
files= []
for i in madrid:
    df = pd.read_csv(i)
    files.append(df)
```

```python
files[5].head()
```

```python
master_frame = pd.concat(files,axis=0, ignore_index=True)
```

```python
master_frame[master_frame['station']==28079035]
```

```python
df_station = master_frame.groupby('station')
df_station1 = df_station.get_group(28079035)
```

```python
df_station1
```

```python
df_station1.isnull().sum()/115656 *100
```

```python
df_station1.drop(['PM25','CH4'],axis=1,inplace= True)
```

```python
df_station1['CO'].fillna(0,inplace = True)
df_station1['NO_2'].fillna(0,inplace = True)
df_station1['SO_2'].fillna(0,inplace = True)
df_station1['BEN'].fillna(df_station1['BEN'].median(), inplace = True)
df_station1['EBE'].fillna(df_station1['EBE'].median(), inplace = True)
df_station1['MXY'].fillna(df_station1['MXY'].median(), inplace = True)
df_station1['NMHC'].fillna(df_station1['NMHC'].median(), inplace = True)
df_station1['NOx'].fillna(df_station1['NOx'].median(), inplace = True)
df_station1['OXY'].fillna(df_station1['OXY'].median(), inplace = True)
df_station1['PM10'].fillna(df_station1['PM10'].median(), inplace = True)
df_station1['PXY'].fillna(df_station1['PXY'].median(), inplace = True)
df_station1['TCH'].fillna(df_station1['TCH'].median(), inplace = True)
df_station1['TOL'].fillna(df_station1['TOL'].median(), inplace = True)
df_station1['NO'].fillna(df_station1['NO'].median(), inplace = True)
```

```python
df_station1.info()
```

```python
from datetime import datetime
df_station1['date']= pd.to_datetime(df_station1['date'], infer_datetime_format = True)
```
