```python
import pandas as pd
import glob
files = glob.glob("C:/Users/uchei/OneDrive/Desktop/csvs_per_year/csvs_per_year**/*.csv")

df = pd.DataFrame()
for f in files:
    csv = pd.read_csv(f, parse_dates=['date'], index_col='date')
    df = df.append(csv)
```

```python
df.head()
```

```python
df.dtypes
```

```python
#Formatting
df['CO'] = df['CO']/1000
df['TCH'] = df['TCH']/1000
df['CH4'] = df['CH4']/1000
```

```python
df.isnull().sum()
```

```python
df_stations = df.groupby('station')
casa_de_campo = df_stations.get_group(28079024)
casa_de_campo.head()
```

```python
casa_de_campo.shape
```

```python
#Percentage of null values contained in a column
casa_de_campo.isnull().sum()/151416*100
```

Casa de Campo contains very few values of CH4. We can drop this gas as we cannot make any useful inference from it in this station.

```python
casa_de_campo = casa_de_campo.drop(['CH4'], axis = 1)
casa_de_campo.head()
```

```python
#Filling null values with medians.
casa_de_campo['BEN'].fillna(casa_de_campo['BEN'].median(), inplace = True)
casa_de_campo['CO'].fillna(casa_de_campo['CO'].median(), inplace = True)
casa_de_campo['EBE'].fillna(casa_de_campo['EBE'].median(), inplace = True)
casa_de_campo['MXY'].fillna(casa_de_campo['MXY'].median(), inplace = True)
casa_de_campo['NMHC'].fillna(casa_de_campo['NMHC'].median(), inplace = True)
casa_de_campo['NO_2'].fillna(casa_de_campo['NO_2'].median(), inplace = True)
casa_de_campo['NOx'].fillna(casa_de_campo['NOx'].median(), inplace = True)
casa_de_campo['OXY'].fillna(casa_de_campo['OXY'].median(), inplace = True)
casa_de_campo['O_3'].fillna(casa_de_campo['O_3'].median(), inplace = True)
casa_de_campo['PM10'].fillna(casa_de_campo['PM10'].median(), inplace = True)
casa_de_campo['PXY'].fillna(casa_de_campo['PXY'].median(), inplace = True)
casa_de_campo['SO_2'].fillna(casa_de_campo['SO_2'].median(), inplace = True)
casa_de_campo['TCH'].fillna(casa_de_campo['TCH'].median(), inplace = True)
casa_de_campo['TOL'].fillna(casa_de_campo['TOL'].median(), inplace = True)
casa_de_campo['PM25'].fillna(casa_de_campo['PM25'].median(), inplace = True)
casa_de_campo['NO'].fillna(casa_de_campo['NO'].median(), inplace = True)

```

```python
#Confirming no null values exist
casa_de_campo.isnull().sum()
```

```python
#Saving cleaned Casa de Campo file
casa_de_campo.to_csv("C:/Users/uchei/OneDrive/Desktop/Gozie/Hamoye Premier Project/Casa_de_Campo_clean.csv", index = False)
```
