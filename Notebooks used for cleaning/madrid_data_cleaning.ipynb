```python
# importing all packages
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from datetime import datetime

%matplotlib inline
```

```python
# loading the each year dataset from 2001-2018 into different pandas dataframe
df_2001 = pd.read_csv('madrid_2001.csv')
df_2002 = pd.read_csv('madrid_2002.csv')
df_2003 = pd.read_csv('madrid_2003.csv')
df_2004 = pd.read_csv('madrid_2004.csv')
df_2005 = pd.read_csv('madrid_2005.csv')
df_2006 = pd.read_csv('madrid_2006.csv')
df_2007 = pd.read_csv('madrid_2007.csv')
df_2008 = pd.read_csv('madrid_2008.csv')
df_2009 = pd.read_csv('madrid_2009.csv')
df_2010 = pd.read_csv('madrid_2010.csv')
df_2011 = pd.read_csv('madrid_2011.csv')
df_2012 = pd.read_csv('madrid_2012.csv')
df_2013 = pd.read_csv('madrid_2013.csv')
df_2014 = pd.read_csv('madrid_2014.csv')
df_2015 = pd.read_csv('madrid_2015.csv')
df_2016 = pd.read_csv('madrid_2016.csv')
df_2017 = pd.read_csv('madrid_2017.csv')
df_2018 = pd.read_csv('madrid_2018.csv')
```

```python
# checking the shape of each dataframe
print(df_2001.shape)
print(df_2002.shape)
print(df_2003.shape)
print(df_2004.shape)
print(df_2005.shape)
print(df_2006.shape)
print(df_2007.shape)
print(df_2008.shape)
print(df_2009.shape)
print(df_2010.shape)
print(df_2011.shape)
print(df_2012.shape)
print(df_2013.shape)
print(df_2014.shape)
print(df_2015.shape)
print(df_2016.shape)
print(df_2017.shape)
print(df_2018.shape)
```

```python
# concatenating all dataframes into a dataframe
df_to_concat=[df_2001,
             df_2002,
             df_2003,
             df_2004,
             df_2005,
             df_2006,
             df_2007,
             df_2009,
             df_2010,
             df_2011,
             df_2012,
             df_2013,
             df_2014,
             df_2015,
             df_2016,
             df_2017,
             df_2018]
df = pd.concat(df_to_concat, axis = 0)
df
```

```python
# checking the columns
df.columns
```

```python
# checking overall information of the dataset
df.info()
```

```python
#converting date into datetime
df['date'] = pd.to_datetime(df['date'], infer_datetime_format=True)
```

```python
# ensuring uniform measurement
df['CO'] = df['CO']/1000
df['TCH'] = df['TCH']/1000
df['CH4'] = df['CH4']/1000
```

```python
# making a copy of the dataset for each station
df_stat1 = df.copy()
df_stat2 = df.copy()
df_stat3 = df.copy()
```

##### 28079004 - Pza. de EspaÃ±a

```python
df_stat1 = df_stat1.loc[(df_stat1['station'] == 28079004)]
df_stat1
```

```python
df_stat1.duplicated().sum()
```

```python
# checking overall information of the dataset df_clean
df_stat1.info()
```

```python
# checking for missing values
df_stat1.isnull().sum()
```

```python
# dropping columns with no values in them


df_stat1.drop(['BEN','EBE','MXY','NMHC','OXY','PXY','TCH','TOL','PM25','CH4'], axis=1, inplace=True)
```

```python
#checking the percentage of missing columns
df_stat1.isnull().sum()/df_stat1.shape[0]*100
```

```python
# treating missing values less than 1 %
df_stat1[['NO_2', 'CO','SO_2']]= df_stat1[['NO_2','CO', 'SO_2']].fillna(0)
```

```python
# treating missing values greater than 1 %
df_stat1[['NOx', 'O_3','PM10']]= df_stat1[['NOx','O_3', 'PM10']].fillna(df_stat1[['NOx','O_3', 'PM10']].median())
```

```python
df_stat1.info()
```

```python
# checking for missing values
df_stat1.isnull().sum()
```

```python
#saving in a csv value
df_stat1.to_csv("Pza de EspaÃ±a.csv", index = False)
```
